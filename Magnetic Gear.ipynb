```python
# Torque vs Air Gap Analysis

This notebook analyzes how torque varies with air gap in a coaxial magnetic gear system.

Objective:
To understand how increasing air gap affects torque transmission due to reduced magnetic coupling.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Air gap values (mm)
air_gap = np.array([1, 2, 3, 4, 5])

# Torque values (Nm)
torque = np.array([11.2, 7.8, 5.1, 3.4, 2.3])
```

```python
x_smooth = np.linspace(min(air_gap), max(air_gap), 100)
y_smooth = np.interp(x_smooth, air_gap, torque)

plt.figure()

plt.plot(air_gap, torque, 'o', label='Data')
plt.plot(x_smooth, y_smooth, '-', label='Trend')

plt.xlabel("Air Gap (mm)")
plt.ylabel("Torque (Nm)")
plt.title("Torque vs Air Gap (Smooth Curve)")

plt.legend()
plt.grid()

plt.show()
```

```python
# Calculate percentage drop
percentage_drop = ((torque[0] - torque) / torque[0]) * 100

for i in range(len(air_gap)):
    print(f"Gap {air_gap[i]} mm → Torque drop: {percentage_drop[i]:.2f}%")
```

```python
## Observations

- Torque decreases as air gap increases.
- The relationship is non-linear.
- Small increases in air gap result in significant reductions in torque.

## Conclusion

Magnetic coupling weakens with increasing distance between rotors, leading to reduced torque transmission.

This supports the concept of controlling torque output by dynamically adjusting the air gap in magnetic gear systems.
```
